# Visualização Gráfica

### LOSS

```python
f = open('loss.txt', "r")


l = []
for i in f.readlines():
    l.append(round(float(i.strip()),2))
print(len(l))
y = l
x = [k for k in range(len(l))]

plt.figure( figsize=(13, 6))  
# plotting the points 
plt.plot(x, y)

plt.xticks([0,200000,400000, 600000, 800000, 1000000])


# naming the x axis
plt.xlabel('Number of timesteps')
# naming the y axis
plt.ylabel('Loss')
  
# giving a title to my graph
plt.title('Train loss results')

# function to show the plot
plt.show()
```

### Rewards

```python
f = open('reward.txt', "r")


l = []
for i in f.readlines():
    l.append(round(float(i.strip()),2))
print(len(l))
y = l
x = [k for k in range(len(l))]

plt.figure( figsize=(13, 6))  
# plotting the points 
plt.plot(x, y, "r-")



# naming the x axis
plt.xlabel('Number of timesteps')
# naming the y axis
plt.ylabel('Reward')
  
# giving a title to my graph
plt.title('Train rewards results')

# function to show the plot
plt.show()
```
